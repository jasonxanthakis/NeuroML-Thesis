```python
import matplotlib.pyplot as plt
from matplotlib import rcParams
import numpy as np

#Extract Data from File
def extract_data(files):
    data = {files[x]:[] for x in range(len(files))}
    for i in range(len(files)):
        filename = files[i]
        data[files[i]] = np.loadtxt(filename)
    
    return data

#Create Combined Plot
def plot_points(data, name, color):
    x_points = data[:, 0]
    y_points = data[:, 1]
    plt.plot(x_points, y_points, color=color, label=name, marker="o")

if __name__ == "__main__":
    #Initiation
    rcParams.update({"font.size": 12})
    plt.figure()

    #Inputs
    filenames = ["Yao2022HumanL23VIP.if.dat", "Yao2022HumanL23SST.if.dat", "Yao2022HumanL23PV.if.dat", "Yao2022HumanL23Pyr.if.dat"]
    names = ["VIP", "SST", "PV", "Pyr"]
    colors = ['#F2CE55', 'green', 'red', 'black']
    
    #Plot Lines
    data = extract_data(filenames)
    for i in range(len(filenames)):
        plot_points(data[filenames[i]], names[i], colors[i])
    
    #Set Up Plot
    plt.xlabel("Input Current (pA)")
    plt.ylabel("Firing Rate (Hz)")
    plt.legend()
    plt.grid()
    plt.savefig("Combined_IF_Graph.png", bbox_inches="tight")
    plt.show()

    for i in range(len(filenames)):
        plt.figure()
        plot_points(data[filenames[i]], names[i], colors[i])
        plt.xlabel("Input Current (pA)")
        plt.ylabel("Firing Rate (Hz)")
        plt.grid()
        plt.savefig("%s_IF_Graph.png"%names[i], bbox_inches="tight")
        plt.show()
```

```python
import matplotlib.pyplot as plt
from pyneuroml import pynml
import numpy as np

lems_file1 = "../delMolino_Adjusted_For_Yao_Model/RateBased.xml"
lems_file2 = "RateBasedSpecifications_high_baseline.xml"
lems_file2 = "RateBasedSpecifications_low_baseline.xml"

populations = ["vip", "sst", "pv", "exc"]
colors = ['yellow', 'green', 'red', 'black']
n_units = 20

def FI_curve(baseline):
    r = {}
    v = {}
    plt.figure()
    for population_idx, population in enumerate(populations):
        lems_file = 'Fig2/LEMS_fISim_%s.xml' %population
        print(lems_file)
        results = pynml.run_lems_with_jneuroml(lems_file, nogui=True, load_saved_data=True, plot=False)

        r[population] = np.zeros((n_units))
        v[population] = np.zeros((n_units))
        for unit in range(n_units):
            pop_unit_r = '%sPop[%d]/r' %(population, unit)
            pop_unit_V = '%sPop[%d]/V' %(population, unit)
            r[population][unit] = results[pop_unit_r][-1]
            v[population][unit] = results[pop_unit_V][-1]

        plt.plot(v[population], r[population], marker='.', color=colors[population_idx])
        
        baseline_lems_file = 'Fig1/LEMS_RateBased_%s_baseline.xml' %baseline
        results_baseline = pynml.run_lems_with_jneuroml(baseline_lems_file, nogui=True, load_saved_data=True, plot=False, verbose=False)

        population_v = '%sPop[0]/V' %population
        population_r = '%sPop[0]/r' %population

        pop = '%sPop' %population
        plt.plot(results_baseline[population_v][1], results_baseline[population_r][1], 'o', label=pop, color=colors[population_idx])

        plt.xlabel('Voltage (V)')
        plt.ylabel('Rate (Hz)')
        plt.xlim(-.060, -.040)
        plt.ylim(0, 60)
        plt.title('{} baseline'.format(baseline))
        plt.legend(loc=(1.04,0))

if __name__ == "__main__":
    FI_curve('low')
    FI_curve('high')
```
